```python
import random
import matplotlib.pyplot as plt

# Number of trials
num_trials = 10000

# Lists to store the minimum and maximum values
min_values = []
max_values = []

for _ in range(num_trials):
    # Roll the die 4 times
    rolls = [random.randint(1, 20) for _ in range(3)]
    # Record the minimum and maximum of the rolls
    min_values.append(min(rolls))
    max_values.append(max(rolls))

# Calculate weights for percentages
weights = [100.0 / num_trials] * num_trials  # Each entry contributes equally to 100%

# Plotting histograms
plt.figure(figsize=(14, 6))

# Histogram of minimum values
plt.subplot(1, 2, 1)
plt.hist(min_values, bins=range(1, 22), weights=weights, align='left', edgecolor='black', color='skyblue')
plt.title('Histogram of Minimum Values')
plt.xlabel('Minimum Value')
plt.ylabel('Percentage (%)')
plt.xticks(range(1, 21))
plt.yticks(range(0, 101, 10))  # Adjust y-ticks to show percentages

# Histogram of maximum values
plt.subplot(1, 2, 2)
plt.hist(max_values, bins=range(1, 22), weights=weights, align='left', edgecolor='black', color='salmon')
plt.title('Histogram of Maximum Values')
plt.xlabel('Maximum Value')
plt.ylabel('Percentage (%)')
plt.xticks(range(1, 21))
plt.yticks(range(0, 101, 10))  # Adjust y-ticks to show percentages

plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Number of trials
num_trials = 1000

# Lists to store the minimum and maximum values
min_values = []
max_values = []

for _ in range(num_trials):
    # Roll the die 4 times
    rolls = [random.randint(1, 20) for _ in range(3)]
    # Record the minimum and maximum of the rolls
    min_values.append(min(rolls))
    max_values.append(max(rolls))

# Calculate weights for percentages
weights = [100.0 / num_trials] * num_trials  # Each trial contributes equally to 100%

# Plotting cumulative histograms
plt.figure(figsize=(14, 6))

# Cumulative histogram of minimum values
plt.subplot(1, 2, 1)
plt.hist(min_values, bins=range(1, 22), weights=weights, cumulative=True, align='left', edgecolor='black', color='skyblue')
plt.title('Cumulative Histogram of Minimum Values')
plt.xlabel('Minimum Value')
plt.ylabel('Cumulative Percentage (%)')
plt.xticks(range(1, 21))
plt.yticks(range(0, 101, 10))  # Adjust y-ticks to show percentages

# Cumulative histogram of maximum values
plt.subplot(1, 2, 2)
plt.hist(max_values, bins=range(1, 22), weights=weights, cumulative=True, align='left', edgecolor='black', color='salmon')
plt.title('Cumulative Histogram of Maximum Values')
plt.xlabel('Maximum Value')
plt.ylabel('Cumulative Percentage (%)')
plt.xticks(range(1, 21))
plt.yticks(range(0, 101, 10))  # Adjust y-ticks to show percentages

plt.tight_layout()
plt.show()
```
